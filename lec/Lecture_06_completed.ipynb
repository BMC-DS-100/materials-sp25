#### Let's look at ages

```python
## Compute the ages of the movie (assuming this is from 2025) and add it as a column
age = 2025 - top_movies.column('Year')
top_movies = top_movies.with_column('age', age)
top_movies
```

Its a good idea to get a sense of the values in our new column. 

Lets look at the min and max ages

```python
max(top_movies.column('age')), min(top_movies.column('age'))
```

```python
top_movies.where("age", 104)
```

```python
top_movies.where("age", 8)
```

```python
min(top_movies.column('age')), max(top_movies.column('age'))
```

```python
top_movies.where('age', 96)
```

### Now let's choose our bins, but we care more about recent movies (make later intervals bigger)

```python
## make an array of our bins
#my_bins = make_array(5, 10, 15, 20, 25, 30, ..., 90) # Add bins here
#my_bins = np.arange(5, 95, 5)
my_bins = make_array(10, 15, 20,25, 30, 35, 50, 110)
my_bins
```

```python
top_movies.bin('age', bins=my_bins)
```

```python
top_movies.where('age', are.below(10)).num_rows
```

```python
top_movies.where('age', are.between(10, 15)).num_rows
```

# Functions

```python
def spread(values):
    spread_val = max(values) - min(values)
    return spread_val
```

```python
def f(s):
    return np.round(s / sum(s) * 100, 2)

```
