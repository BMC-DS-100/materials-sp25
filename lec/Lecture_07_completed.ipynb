#### Multiple Arguments & Default Arguments

```python
def double_and_multiply(arg1, arg2):
    """
    Computes and returns the value of 2*arg1 * 2*arg2
    """
    return double(arg1) * double(arg2)
```

```python
help(double_and_multiply)
```

```python
double_and_multiply(3, 6)
```

```python
def double_and_multiply2(arg1, arg2=2):
    """
    Computes and returns the value of 2*arg1 * 2*arg2

      - arg2: default value is 2
    """
    return double(arg1) * double(arg2)
```

# `Apply`ing functions to Tables

```python
def cut_off_at_a_billion(x):
    '''
    Returns x if x is less than 1e9, otherwise returns 1e9
    '''
    if x > 1e9:
        return 1e9
    return x
```

```python
cut_off_at_a_billion(1)
```

```python
cut_off_at_a_billion(1e10)
```

```python
top_movies = Table.read_table('top_movies_2017.csv')
top_movies
```

### Apply billion cut off function to a Column and add it to the table

```python
gross_adj_cut = top_movies.apply(cut_off_at_a_billion, 'Gross (Adjusted)')
top_movies.with_column('Gross (Adjusted) 1B max', gross_adj_cut)
```

```python
top_movies
```

```python
top_movies = top_movies.with_column('Gross (Adjusted) 1B max', gross_adj_cut)
top_movies
```

```python
# let's compute the mean height of children in this range
nearby = galton.where('midparentHeight', are.between(67.5, 68.5))
nearby_mean = nearby.column('childHeight').mean()
nearby_mean
```

```python
# Lets schow the mean child height 
galton.scatter('midparentHeight', 'childHeight')
plots.plot([67.5, 67.5], [50, 85], color='red', lw=2)
plots.plot([68.5, 68.5], [50, 85], color='red', lw=2)
plots.scatter(68, nearby_mean, color='red', s=50);
```

```python
# Now let's right a function to make this prediction in general
```

```python
def predict_height(parentsMidHeight):
    #find the instancews in our data where the parentmidheight is within a 0.5 range of the given height
    nearby = galton.where('midparentHeight', are.between(parentsMidHeight - 0.5, parentsMidHeight + 0.5))
    nearby_mean = nearby.column('childHeight').mean()
    return nearby_mean
```

```python
predict_height(68)
```

```python
predict_height(74)
```

```python
galton.apply(predict_height, 'midparentHeight')
```

```python
galton = galton.with_column('predictedHeight', galton.apply(predict_height, 'midparentHeight'))
galton
```

```python
galton.select(
    'midparentHeight', 'childHeight', 'predictedHeight').scatter('midparentHeight')
```

# Pivot

Use pivot here!
